*   **Import the OS module --> to list files and access directories**
*   **Import the annotate function to manually label data**
*   **Import display and image functions to show images in Jupyet Notebook**

```python
import os
from pigeon import annotate
from IPython.display import display, Image
```

```python
#Defines the path to the folder with the images
folder_path = '/content/Boots and Sneakers Dataset'
#Stores all the file names
image_names = os.listdir(folder_path)
#Contains the full paths to each image by combining the folder path with each image file name.
images_paths = [os.path.join(folder_path, image_name) for image_name in image_names]
```

```python
len(images_paths)
```

```python
label_options = ["boots", "sneakers"]
```

```python
annotations = annotate(images_paths, options=label_options, display_fn=lambda filename:display(Image(filename)))
```

```python
annotations
```

```python
import pandas as pd
df_annotation = pd.DataFrame(annotations,columns=['filename', 'label'])
df_annotation
```

```python
df_annotation.to_csv('boots_and_sneakers_annotation.csv', index = False)
```
